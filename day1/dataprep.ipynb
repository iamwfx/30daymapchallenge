```python
import pandas as pd
import numpy as np
import os 
```

Read in the data

```python
litigations = pd.read_csv('Data/Housing_Litigations.csv')
```

```python
litigations.columns
```

```python
litigations[['CaseType','CaseStatus',
       'OpenJudgement', 'FindingOfHarassment', 'FindingDate', 'Penalty',
       'Respondent',]]
```

```python
litigations.groupby('CaseStatus').count()
```

```python
litigations.groupby('CaseType').count()
```

```python
litigations['CaseStatusNew'] = litigations['CaseStatus'].apply(lambda x: x.replace('-',',').split(',')[0].replace('D ','D'))
```

```python
litigations['CaseStatusNew'].unique()
```

Turn date into datetime

```python
litigations['CaseOpenDate']=pd.to_datetime(litigations['CaseOpenDate'])
litigations['FindingDate']=pd.to_datetime(litigations['FindingDate'])
```

```python
litigations['YearCaseOpen'] = litigations['CaseOpenDate'].dt.year
litigations['MonthCaseOpen'] = litigations['CaseOpenDate'].dt.month
```

```python
litigations.to_csv('Data/Housing_Litigations_NEW.csv')
```

```python
import geopandas as gpd
from shapely.geometry import Point
```

```python
litigations_gpd = gpd.GeoDataFrame(litigations,geometry=[Point(xy) for xy in zip(litigations.Longitude, litigations.Latitude)],crs="EPSG:4326")
```

```python
litigations_gpd.to_file('Data/Housing_Litigations_NEW.geojson',driver='GeoJSON')
```
